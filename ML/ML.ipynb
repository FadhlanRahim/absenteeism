```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('preprocesseddf.csv')
df.head()
```

# Targets

First, we must decide that is this a classification or regression problem, if we refer the the absent hour time feature, it is filled with the amount of hour certain employee has been absent, if we are going to predict that, this mean that this is a regression problem.

Alternatively, we can turn this into a classification, from the amount of time absent, we can decide how many is excessive and how many is normal, many way to decide this, but to make it simple, lets refer to the median.

Lets decide that this is a classification problem so we have to turn the amount of hours into classes of excesive or not.

```python
df.describe()
```

```python
df.iloc[:, -1].median()
```

From that, we know that majority of amount of absent time is 3 Hours. lets decide that this is the cut-off point from normal amount and excesive amount.

```python
df['Target'] = np.where(df.iloc[:, -1] <= 3, 0, 1)
df['Target']
```

```python
df
```

Target :
* 0 = No absent or normal amount of absent
* 1 = Excessive

```python
df['Target'].value_counts(normalize=True)
```

```python
#drop absent time hours and Year
df2 = df.drop(['Absenteeism Time in Hours', 'Year'], axis =1)
df2
```

Separate inputs and target(output)

# Splitting

To train ML model, it is advised to not train it with the whole data, instead make 2 (sometimes 3) sets so we have a set for training the model, then test the model with another set that is not used to train, in other word, the model has not seen this portion of the data and will make a prediction on this data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df2.drop('Target', axis=1), df2.Target, test_size = 0.25, stratify = df2.Target)
```

# Standardize

Notice that our numerical data is varied by ranges, before we use this data for our machine learning model training, we have to standardize it so feature with huge number will not affect the model to be biased toward that feature.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

We scale the test set using the training set attribute, hence, no fitting.

# Modeling: Logistic Regression

Finally, we come to the actual ML part, because we already decide that we want to predict that the absent amount is Excessive(noted by '1') OR not(noted by '0'). we decide to use Logistic Regression as our estimator

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train)
```

```python
logreg.score(X_train_scaled, y_train)
```

Our model accuracy is : 71%
but this is on training set, lets test it on test set

## Intercept and Coeficient

```python
logreg.intercept_
```

```python
logreg.coef_
```

```python
featurename = X_train.columns.values
```

```python
sumtable = pd.DataFrame(columns = ['Feature Name'], data = featurename)
sumtable['Coef'] = np.transpose(logreg.coef_)
sumtable
```

```python
sumtable.index = sumtable.index+1
sumtable.loc[0] = ['Intercept', logreg.intercept_[0]]
```

```python
sumtable.sort_index(inplace=True)
sumtable
```

# Interpreting Coefficients

```python
sumtable['Odds_ratio'] = np.exp(sumtable.Coef)
sumtable
```

```python
sumtable.sort_values('Odds_ratio', ascending=False)
```

Interpretation : for each increased unit in features, the odds will be multiplied by the value of odds_ratio.

For example : if we have an odds of (our intercept is 0.83, lets round it up for simplification)5:1, and our employee is absent with reason3, the odds is now 8.12:1, increasing the odds of absent being excessive(from 5:1), on the other hand, if employee is absent with reason4, the odds is now 3.45:1 decreasing the odds of absent being excessive(from 5:1). Same rules applied for other features including numerical but remember that we standardize the numerical feature, so, for example rather than for each unit, for example 'Dialy Work Load Average', its standardize unit of 'Dialy Work Load Average'.

Can the model be improved? What can we do?
well yes, when we standardize the data, we also standardize the dummies which only contains 0 and 1, in other words its not actually a continuous features, we should not do that in the first place.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df2.drop(['Target', 'Reason4'], axis=1), df2.Target, test_size = 0.25, stratify = df2.Target)
```

```python
from sklearn.compose import make_column_transformer
df2.columns
```

```python
#Select Features to be transformed (leave all categorical features)
coltotrans = ['Month', 'Day_of_the_week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Children',
       'Pets']
ct = make_column_transformer([StandardScaler(), coltotrans], remainder='passthrough')
X_train_scaled = ct.fit_transform(X_train)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = ['Month', 'Day_of_the_week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Children',
       'Pets', 'Reason1', 'Reason2', 'Reason3', 'Education'])
X_train_scaled
```

The categorical feature is untouched, transform our test set.

```python
X_test_scaled = ct.transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = ['Month', 'Day_of_the_week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Children',
       'Pets','Reason1', 'Reason2', 'Reason3', 'Education'])
X_test_scaled
```

```python
# Retrain
logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train)
```

```python
logreg.score(X_train_scaled, y_train)
```

```python
# reinterpret
featurename = X_train_scaled.columns.values
sumtable = pd.DataFrame(columns = ['Feature Name'], data = featurename)
sumtable['Coef'] = np.transpose(logreg.coef_)
sumtable.index = sumtable.index+1
sumtable.loc[0] = ['Intercept', logreg.intercept_[0]]
sumtable.sort_index(inplace=True)
sumtable['Odds_ratio'] = np.exp(sumtable.Coef)
sumtable
```

We leave reason4 to be our baseline, reason4 are absent with no reason or light reason(dental, medical consult, etc).

So to re-interpret :
1. if employee absent with reason3 (injury and/or poisoning), that employee will 12.5 times likely to take excessive amount of absent compared to employee who absent with reason4, same with other reason with their respectiv odds ratio.
2. an employee with college-grad above will 0.98 times likely to take excessive absent
3. for numerical data, for each STANDARDIZE unit of that value, not unit of that values. so its not intepretable.

# More Metrics

Accuracy alone is not enough metrics to evaluate the model, enter confusion matrix, we will take a look at model's each prediction compared to the actual label.

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, pred_class)
```

* The Top left is prediction 0, with actual label 0, A true Negatives
* The Bottom Right is prediction 1, with actual label 1, A true Positives
* The top right is prediction 1 with actual label 0, a false positive
* The bottom left is prediction 0 with actual label 1, a false negative

```python
print(classification_report(y_test, pred_class))
```

* Precision : How many positive prediction is correct prediction? we will have lower precision if we have more false positive. our model has 0.73 precision on positive class, that means, when it predict an employee taking an exesive amount of absent, it is correct 73% of the time
* Recall  : How many actual positive was indentified correctly? we will have lower recall if we have more false negative. our model has 0.56 recall on positive class, that means, the model correctly identified 56% of all employee on positive class(excessive amount of absent time)

In short, model with no false positive has precision of 1, and model with no false negative has recall of 1. If a business is impacted with false negative, in this case an employee that will take excessive amount of absent is predicted 0(will not take excessive amount of absent) the business might need a model with high recall, to do this we can adjust the prediction threshold, by default that threshold is 0.5.

```python
adj_pred_class = np.where(pred_proba[:, 1] < 0.4, 0, 1)
#prediction more than or equal to 0.3 probability will be classified as positive class, will take excessive amount of time of absent
print(classification_report(y_test, adj_pred_class))
```

Notice how the precision is now reduced, because the lower threshold, we now have more false positive.

Before saving the model, because to train the model we have to scale the data first, in other words, that scaling process is part of the model. for that we have to combine the scaling process with our logisticregression model, enter Pipeline.

```python
#scaler is saved as ct and logisticregression is saved as logreg
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(ct, logreg)
```

```python
# test the model again, now with unscaled data because it will go through the included scaling process
pipe.score(X_test, y_test)
```
